## Variables

No need to declare variable types/scope/privacy in Python. Python will detect whatever type you want your variable to be automatically!

```python
# Declare a variable
x = 5
y = "Hello, BDAA!"
z = False
l = [x, y, z]
print(x, y, z)
print(l)
```

If you'd like to explicitly set a variable type, use the base *str(), int(), float(), and bool()* functions.

```python
# Declare a variable, explicitly
x = str(5)
y = float(5)
z = str(False)
print(x, y, z)
```

Since this may be important for our class.. <br/>
If you'd like to build a string using a variable you have, make sure you declare that variable as a string.

```python
# Build a string
x = 74
y = "The Tom W. Davis Clocktower is " + str(x) + " feet high."
print(y)
```

## Conditionals, and loops

Here's how if statements work in Python:

```python
x = 5
if x == 6:
    print("This shouldn't print..")
elif x == 2012079474:
    print("This definitely shouldn't print..")
else:
    print("Hooray!")
```

And Loops! (mmmm.. gyros) <br/> <br/>

Here's three ways of doing *roughly* the same thing:

```python
x = [1,2,3,4,5]

for number in x:
    print(number)
```

```python
x = [1,2,3,4,5]

for i in range(len(x)):
    print("Element at index " + str(i) + ": " + str(x[i]))
```

```python
x = [1,2,3,4,5]

for i, number in enumerate(x):
    print("Element at index " + str(i) + ": " + str(number))
```

## List Comprehensions

List comprehensions are an incredibly powerful, and eloquent perk of Python. To prove it, let's take a look at an example:

```python
x = [1, 2, 3, 4, 5, 6]
```

How would you multiply each element in this list/array by, say, 5 in most other languages? You might do something like this:

```python
y = [0, 0, 0, 0, 0, 0]
for i in range(len(x)):
    y[i] = x[i] * 5

print(y)
```

In Python, you can do this operation (among <b>many</b> others) in one beautiful line of code:

```python
y = [num*5 for num in x]
print(y)
```

## String Operations

String operations will be *extremely* important in this workshop, so listen up!

```python
x = "   \n \n \r This-is-a-very-dirty    \n\n-string.    "
print(x)
```

Let's clean this string up!

```python
# strip() removes all leading and trailing whitespaces
x = x.strip()
print(x)
```

```python
# replace() will match any substring and replace it with whatever you'd like
x = x.replace('\n', '')
print(x)
x = x.replace('  ', '')
print(x)
```

```python
# split() will convert your string into a list, seperated by whatever character/substring you'd like
x = x.split('-')
print(x)
```

```python
# <string>.join(<variable>) will join your <variable> (a list type) back into a string, joined by your <string>
x = ' '.join(x)
print(x)
x = x.replace('dirty', 'clean')
print(x)
```
